```python
from langchain_community.tools import DuckDuckGoSearchRun
from typing import List
import re
from collections import Counter

ddg = DuckDuckGoSearchRun()

def extract_keywords(query: str) -> List[str]:
    
    stop_words = set(["what", "is", "the", "of", "and", "for", "to", "in", "a", "an", "by", "with"])
    words = re.findall(r'\w+', query.lower())
    keywords = [word for word in words if word not in stop_words]
    return keywords

def ddg_search(input_text:str):
    
    result = ddg.run({"query": input_text, "num_results": 5})

    return result

def evaluation(results: str, query: str) -> str:


    if not results or "Error" in results:
        return "No valid results found. Try a different query."
    
    # Extract keywords from the query
    keywords = extract_keywords(query)
    
    # Keyword Relevance
    keyword_score = sum(1 for word in keywords if word.lower() in results.lower())
    max_keyword_score = len(keywords) or 1
    relevance_score = (keyword_score / max_keyword_score) * 5  # Scale to 5

    # Length Evaluation (Up to 5 points for sufficient length)
    length_score = min(len(results) / 500, 5)
    
    # Readability Evaluation (Basic structure analysis)
    sentence_count = results.count('.') + results.count('!') + results.count('?')
    word_count = len(re.findall(r'\w+', results))
    readability_score = min((sentence_count / max(word_count / 100, 1)) * 5, 5)
    
    # Final Score Calculation
    final_score = relevance_score + length_score + readability_score
    final_score = min(final_score, 10)  # Cap at 10
    
    return f"Evaluation Score: {final_score:.2f}/10"

query1 = "What is the capital of China?"
query2 = "What are the top research papers on maximising hypertrophy in muscle building in 2023?"

simple_search = ddg_search(query1)
complex_search = ddg_search(query2)

print(f"Query: {query1}")
print(simple_search)
print("Evaluation: ", evaluation(simple_search, query1))

print(f"\nQuery: {query2}")
print(complex_search)
print("Evaluation: ", evaluation(complex_search, query2))

        
```
